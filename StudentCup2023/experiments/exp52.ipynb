```python
import numpy as np
import pandas as pd
import polars as pl
```

```python
class CFG:
    seed = 42
    save_dir = "/content/drive/MyDrive/Colab Notebooks/signate2023/exp/"
    data_dir = "/content/drive/MyDrive/Colab Notebooks/signate2023/data/"
    filename = "exp00052"
```

```python

filenames = [f"exp000{f}" for f in range(38, 45)] + [f"exp000{f}" for f in range(46, 49)]

# oof_df
for i, filename in enumerate(filenames):
    path = CFG.save_dir+f"kun_{filename}_oof_pred.csv"
    if i == 0:
        oof_df = pl.read_csv(source=path).rename({"oof_pred":f"kun_pred_{i}"})
    else:
        oof_df = oof_df.join(pl.read_csv(source=path).rename({"oof_pred":f"kun_pred_{i}"}),
                             on="id", how="left")
oof_df.write_csv(CFG.save_dir+f"kun_{CFG.filename}_oof_pred.csv")

# test_df
for i, filename in enumerate(filenames):
    path = CFG.save_dir+f"kun_{filename}.csv"
    if i == 0:
        test_df = pl.read_csv(source=path, has_header=False).rename({"column_1":"id", "column_2":f"kun_pred_{i}"})
    else:
        test_df = test_df.join(pl.read_csv(source=path, has_header=False).rename({"column_1":"id", "column_2":f"kun_pred_{i}"}),
                             on="id", how="left")
test_df.write_csv(CFG.save_dir+f"kun_{CFG.filename}.csv")
```
